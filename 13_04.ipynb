```python
# Scalar operations on numpy array
import numpy as np
a = np.array([1,2,3,4,5])
print(a)
print(a + 1)
print(a - 1)
print(a * 2)
```

```python
print(a * a)
print(a.dot(a))
```

```python
a = np.array([1,1,1,1,0,1], dtype=bool)
print(np.all(a))
print(np.any(a))
```

```python
a = np.array([1,1,1,1,1,1], dtype=bool)
print(np.all(a))
print(np.any(a))
```

```python
a = np.zeros(10)
print(np.all(a))
print(np.all(a == 0))
```

```python
a = np.array([1,2,3,4,5])
b = np.array([9,10,11,12,13])
c = np.array([3,4,5,6,7])

np.all((a <= b) & (a == c))
```

```python
# Transcedental functions
a = np.arange(6).reshape(3, 2)
print(np.sin(a))
print("\n")
print(np.cos(a))
print("\n")
print(np.tan(a))
print("\n")
print(np.exp(a))
print("\n")
print(np.log(a))
```

```python
# Computing sum
a = np.arange(6).reshape(2, 3)
print(a)
print(np.sum(a))
print(np.sum(a, axis=0))
print(np.sum(a, axis=1))
print(np.sum(a, axis=-1))
```

```python
a = np.array([1, 3, 5, 7, 2])
print(a)
print(a.max())
print(a.min())
print(a.argmax()) # Index of max value
print(a.argmin()) # Index of min value
```

```python
# Statistical Operations
a = np.array([1, 2, 3, 4, 5])
b = np.array([[1, 2, 4], [7, 7, 7]])
print(np.mean(a))
print(np.mean(b))
print(np.median(a))
print(np.median(b, axis=1))
print(a.std())
print(b.std(axis=0))
```
