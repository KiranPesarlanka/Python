```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [10, 11, 11, 9])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph - 1')
plt.grid()
plt.show()
```

```python
# If odd number of plots are given as input, the odd one will be considered as y-axis
plt.plot([1, 2, 3, 4], [10, 11, 11, 9], 'b--', [2, 3, 4, 8])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph - 2')
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 11, 11, 9], [2, 3, 4, 8], [0, 1, 1, 2])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph - 3')
plt.grid()
plt.show()
```

```python
lines = plt.plot([1, 2, 3, 4], [10, 11, 11, 9], 'bv', [2, 3, 4, 8], [0, 1, 1, 2], 'b--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.setp(lines[0], linewidth='5')
plt.setp(lines[1], linewidth='10')
print(plt.setp.__doc__)
plt.title('Graph - 4')
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(5)
plt.plot(t, t**2, 'r-', label='t**2')
plt.plot(t, t**3, 'bo', label='t**3')
plt.plot(t, t**4, 'g--', label='t**4')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph - 5')
plt.legend()
plt.grid()
plt.show()
```

# Matplotlib
It is an inbuilt python library which acts as an alternative for MatLab in python.
The pyplot package present in the matplotlib library provides us with many inbuilt functions inorder to perform data visualisation by plotting graphs such as scatter plots, histograms, pie charts, etc.,
It is a very powerful data visualisation tool.

```python
t1 = np.arange(0., 5., 0.5)
t2 = np.arange(0., 5., 0.02)

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

plt.figure(1)

plt.subplot(211)
plt.plot(t1, f(t1), 'r--')
plt.grid()

plt.subplot(212)
plt.plot(t2, f(t2), 'b--')
plt.grid()

plt.figure(2)
plt.subplot(221)
plt.plot(t1, f(t1), 'r--')
plt.grid()

plt.subplot(222)
plt.plot(t2, f(t2), 'b--')
plt.grid()

plt.show()
```

```python
# Excercise have 2 figures, 3 graphs in each figure
import numpy as np
import matplotlib.pyplot as plt

t1 = np.arange(0., 5., 0.5)
t2 = np.arange(0., 5., 0.02)

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

plt.figure(1)

plt.subplot(311)
plt.plot(t1, f(t1), 'r--')
plt.grid()

plt.subplot(312)
plt.plot(t2, f(t2), 'b--')
plt.grid()

plt.subplot(313)
plt.plot(t2, f(t2), 'm--')
plt.grid()

plt.figure(2)
plt.subplot(231)
plt.plot(t1, f(t1), 'y--')
plt.grid()

plt.subplot(232)
plt.plot(t2, f(t2), 'k--')
plt.grid()

plt.subplot(233)
plt.plot(t2, f(t2), 'c--')
plt.grid()

plt.show()
```
